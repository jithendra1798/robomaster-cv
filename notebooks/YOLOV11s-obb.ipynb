```python
import os
from pathlib import Path
from ultralytics import YOLO

PROJECT_ROOT = Path.cwd().parent  # notebooks/ → robomaster-cv/
DATA_DIR = PROJECT_ROOT / 'data/mergeRM'
DATASET_YAML = str(PROJECT_ROOT / 'configs/data.yaml')

os.listdir(DATA_DIR)
```

```python
IMG_DIR = DATA_DIR / 'images'
LABEL_DIR = DATA_DIR / 'labels'

print(os.listdir(IMG_DIR))
print(os.listdir(LABEL_DIR))
```

```python
for dataset in os.listdir(IMG_DIR):
    print(f'{dataset.capitalize()}:', len(os.listdir(IMG_DIR / dataset)), 'and', len(os.listdir(IMG_DIR / dataset)))
```

```python
model = YOLO('yolo11s-obb.pt')

hyp_search_space = {
    'lr0': (1e-5, 1e-1),
    'momentum': (0.6, 0.98),
    'weight_decay': (0.0, 0.001),
    'box': (5, 10),
    'mosaic': (0.0, 1.0)
}

!nvidia-smi
```

```python
%%time

model.tune(
    data=DATASET_YAML,
    epochs=10,
    iterations=20,        # Keep this low for testing; increase for serious optimization
    optimizer='AdamW',
    batch=64,
    imgsz=640,            # Standard image size for YOLO
    device=0,
    name="obb_tune_run",  # Name of the folder where results are saved
    space=hyp_search_space,
    # augmentations from benchmark
    mixup=0.1,
    degrees=15.0,
    shear=2.0,
    perspective=0.0001,
)
```

```python
%%time
import yaml

# Auto-load best hyperparameters from tuning
tune_results_path = "/home/jupyter-robomaster/jithendra/notebooks/runs/obb/obb_tune_run/best_hyperparameters.yaml"

with open(tune_results_path, 'r') as f:
    best_hyps = yaml.safe_load(f)

print("Best hyperparameters from tuning:")
for k, v in best_hyps.items():
    print(f"  {k}: {v}")

model.train(
    data=DATASET_YAML,
    epochs=200,
    optimizer='AdamW',
    batch=64,
    imgsz=640,
    device=0,
    name='obb_final_model',
    exist_ok=True,
    patience=30,
    cos_lr=True,
    amp=True,
    cache='disk',
    # Tuned parameters (auto-loaded)
    lr0=best_hyps.get('lr0', 0.001),
    momentum=best_hyps.get('momentum', 0.9),
    weight_decay=best_hyps.get('weight_decay', 0.001),
    box=best_hyps.get('box', 7.5),
    mosaic=best_hyps.get('mosaic', 1.0),
    # Hardcoded augmentations
    mixup=0.1,
    degrees=15.0,
    shear=2.0,
    perspective=0.0001,
)
```

## Inference

```python
best_model_path = "/home/jupyter-robomaster/jithendra/notebooks/runs/obb/obb_final_model/weights/best.pt"

best_model = YOLO(best_model_path)

# Validation split
val_metrics = best_model.val(data=DATASET_YAML, split='val', device=0)

print(f"\nmAP50:     {val_metrics.box.map50:.4f}")
print(f"mAP50-95:  {val_metrics.box.map:.4f}")
print(f"Precision: {val_metrics.box.mp:.4f}")
print(f"Recall:    {val_metrics.box.mr:.4f}")
```

```python
test_metrics = best_model.val(data=DATASET_YAML, split='test', device=0)

print(f"\nmAP50:     {test_metrics.box.map50:.4f}")
print(f"mAP50-95:  {test_metrics.box.map:.4f}")
print(f"Precision: {test_metrics.box.mp:.4f}")
print(f"Recall:    {test_metrics.box.mr:.4f}")
```
