```python
IMG_DIR = DATA_DIR / 'images'
LABEL_DIR = DATA_DIR / 'labels'

print(os.listdir(IMG_DIR))
print(os.listdir(LABEL_DIR))
```

```python
model = YOLO('yolo11l.pt')

!nvidia-smi
```

```python
%%time

model.train(
    data = DATASET_YAML,
    epochs = 100,
    weight_decay = 0.001,
    box = 6.22111,
    imgsz = 640,
    name = 'baseline_v11l'
)
```
